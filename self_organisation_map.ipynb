```python
import numpy as np
```

```python
data=np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
```

```python
np.random.seed(0)
w=np.random.rand(2,2)
lr=0.1
epochs=10
```

```python
print(w)
```

```python
for i in range(epochs):
    for x in data:
        distances=np.linalg.norm(w-x,axis=1)
        winner=np.argmin(distances)
        w[winner]=w[winner]+lr*(x-w[winner])
```

```python
print("Final Cluster Weights: \n",w)
```

```python
test_val=[[0.1,0.1]]
result=np.argmin(np.linalg.norm(w-test_val,axis=1))
```

```python
print(f"Input {test_val} belongs to Cluster {result}")
```
