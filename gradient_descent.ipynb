```python
n=int(input("Enter the number of input neurons: "))
x=[]
w=[]
for i in range(0,n):
    val1=float(input(f"Enter the value of neuron {i+1}: "))
    x.append(val1)
    val2=float(input(f"Enter the value of weight for neuron {i+1}: "))
    w.append(val2)
b=float(input("Enter the value of bias: "))
print(x)
print(w)
lr=0.1
```

```python
y_target=[]
for i in range(n):
    val3=float(input(f"Enter the target value of neuron {i+1}: "))
    y_target.append(val3)
print(y_target)
```

```python
grads=[]
```

```python
for i in range(len(x)):
    sum_y=(x[i]*w[i])+b
    error=sum_y-y_target[i]
    gradient=2*error*x[i]
    grads.append(gradient)
    
```

```python
print(grads)
```

```python
def batch_gd(grads,w,lr):
    batch=[]
    avg_grad=sum(grads)/len(grads)
    for i in range(len(w)):
        w_batch=w[i]-(lr*avg_grad)
        batch.append(w_batch)
    return batch
```

```python
print("Batch GD:",batch_gd(grads,w,lr))
```

```python
def sgd(grads,w,lr):
    s_gd=[]
    for i in range(len(grads)):
        val=w[i]-(lr*grads[i])
        s_gd.append(val)
    return s_gd
```

```python
print("Stochastic GD:", sgd(grads,w,lr))
```

```python
def mini_batch(grads,w,lr,size=3):
    mini_gd=[]
    current_w=w.copy()
    for i in range(0,len(grads),size):
        batch=grads[i:i+size]
        avg_g=sum(batch)/len(batch)
        for j in range(i,min(i+size,len(w))):
            current_w[j]=current_w[j]-(lr*avg_g)
            mini_gd.append(current_w[j])
        return mini_gd
```

```python
print("Mini-Batch GD:", mini_batch(grads,w,lr,size=3))
```
