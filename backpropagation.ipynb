```python
from math import e as e
```

```python
x=list(map(float,input ("Enter the input values separated by a space: ").split()))
```

```python
y=list(map(float,input("Enter the target value: ").split()))
```

```python
w=list(map(float,input("Enter the initial weights: ").split()))
```

```python
b=float(input("Enter the bias: "))
lr=float(input("Enter the learning rate: "))
```

```python
def sigmoid(x):
    return 1/(1+pow(e,-x))
```

```python
def forward_pass(x,w,b):
    sum_y=sum(x[i]*w[i] for i in range(len(x)))+b
    return sigmoid(sum_y)
```

```python
def backward_pass(y,y_act,x,w,b,lr):
    err=y[0]-y_act
    delta=err*(y[0]*(1-y[0]))
    new_w=[w[i]+lr*delta*x[i]for i in range(len(w))]
    new_b=b+lr*delta
    return new_w,new_b
```

```python
print(f"Initial prediction: {forward_pass(x,w,b)}")
```

```python
y_initial=forward_pass(x,w,b)
new_weights,new_bias=backward_pass(y,y_initial,x,w,b,lr)
print(f"Updated weights: {[round(nw,4) for nw in new_weights]}")
```

```python
y_final=forward_pass(x,new_weights,new_bias)
print(f"Final prediction: {round(y_final,4)}")
```
