```python
from math import e as e
```

```python
def activation(x):
    return 1/(1+pow(e,-x))
```

```python
n=int(input("Enter the number of input neurons"))
```

```python
x=[]
for i in range(n):
    val=float(input("Enter the value of input neuron {i+1}"))
    x.append(val)
print(x)
```

```python
w=[]
for i in range(n):
    val=float(input("Enter the weight of input neuron {i+1}"))
    w.append(val)
print(w)
```

```python
bias=0.5
lr=10.0
y_target=0.7
```

```python
def forward_pass(x,w,bias):
    sum_y=sum(x[i]*w[i] for i in range(len(x)))+bias
    return activation(sum_y)
```

```python
y_initial=forward_pass(x,w,bias)
print("Initial prediction: ",forward_pass(x,w,bias))
```

```python
def backward_pass(y_target,y_actual,x,w,bias,lr):
    error=y_target-y_actual
    delta=error*y_initial*(1-y_initial)
    new_w=[(w[i]+lr*delta*x[i]) for i in range(len(w))]
    new_b=bias+lr*delta
    return new_w,new_b
```

```python
new_weights,new_bias=backward_pass(y_target,y_initial,x,w,bias,lr)
```

```python
print(f"Updated weights: {[round(nw,4) for nw in new_weights]}")
```

```python
y_result=forward_pass(x,new_weights,new_bias)
print(f"Updated result: {round(y_result,4)}")
```
