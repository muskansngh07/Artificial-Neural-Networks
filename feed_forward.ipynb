```python
from math import e as e
```

## Input Layer

```python
n=int(input("Enter the number of neurons in the input layer: "))
x=list(map(float,input("Enter the value of input neurons separated by a space: ").split()))
```

## Processing the hidden layer

```python
h=int(input("Enter the number of neurons in the hidden layer: "))
hidden=[]
```

```python
k=0
for i in range(h):
    w=list(map(float,input("Enter the weight of input neurons separated by a space: ").split()))
    b=float(input(f"Enter the bias for this neuron: "))
    sum=sum+(x[k]*w[k])+b
    hidden.append(1/(1+e**(-sum)))
    k+=1
```

```python
print(hidden)
```

## Processing the output layer

```python
o=int(input("Enter the number of neurons in the output layer: "))
output=[]
```

```python
k=0
for i in range(o):
    w=list(map(float,input("Enter the weight of hidden layer neurons separated by a space: ").split()))
    b=float(input(f"Enter the bias for this neuron: "))
    sum=sum+(hidden[k]*w[k])+b
    output.append(1/(1+e**(-sum)))
    k+=1
```

```python
print(output)
```
