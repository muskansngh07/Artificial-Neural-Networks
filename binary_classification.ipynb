```python
from math import e as e
```

## Input Layer

```python
n=int(input("Enter the number of neurons in the input layer: "))
x=list(map(float,input("Enter the value of input neurons seperated by a space: ").split()))
```

## Hidden Layer

```python
h=int(input("Enter the number of neurons in the hidden layer: "))
```

```python
hidden=[]
k=0
result=0
for i in range(h):
    w=list(map(float, input("Enter the weight of incoming neurons: ").split()))
    b=float(input("Enter the bias for this neuron: "))
    result+=(x[k]*w[k])+b
    hidden.append(1/(1+e**(-result)))
```

```python
print(x)
```

```python
print(hidden)
```

## Output layer

```python
o=2
```

```python
output=[]
k=0
result=0
for i in range(o):
    w=list(map(float, input("Enter the weight of incoming neurons: ").split()))
    b=float(input("Enter the bias for this neuron: "))
    result+=(hidden[k]*w[k])+b
    output.append(1/(1+e**(-result)))
    k+=1
```

```python
print(output)
```

```python
output1=output[0]
output2=output[1]
print(f"Probabaility of Class 1 is {output1}")
print(f"Probability of Class 2 is {output2}")
```

```python
predicted=1 if (output[0]>output[1]) else 2
print(f"The final prediction is Class {predicted}")
```
