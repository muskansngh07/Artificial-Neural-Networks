```python
from math import e as e ## so that e isn't treated as a variable name 
```

## Take inputs

```python
x=list(map(float,input("Enter the value of input neurons separated by spaces: ").split()))
w=list(map(float,input("Enter the value of weight of input neurons separated by spaces: ").split()))
sum_y=0
```

```python
if(len(x)!=len(w)):
    print("Invalid choice!")
```

## Calculate the weighted sum

```python
print(x)
print(w)
```

```python
for i in range(0,len(x)):
    sum_y=sum(x[i]*w[i] for i in range(len(x)))
print(sum_y)
```

## Sigmoid activation function

```python
def sigmoid(x):
    print(f"Sigmoid activation function: {1/(1+e**(-x))}")
```

## Step activation function

```python
def step(x):
    if(x>=0):
        print("Step activation function: 1")
    else:
        print("Step activation function: 0")
```

## tanh activation function

```python
def tanh(x):
    num=(e**x)-e**(-x)
    den=(e**x)+e**(-x)
    print(f"tanh activation function: {num/den}")
```

## Sign activation function

```python
def sign(x):
    if(x>0):
        print("sign activation function: 1")
    elif(x==0):
        print("sign activation function: 0")
    else:
        print("sign activation function: -1")
```

## Linear activation function

```python
def linear(x):
    print(f"Linear activation function: {x}")
```

## ReLU activation function

```python
def relu(x):
    print(f"ReLU activation function: {max(0,x)}")
```

## Leaky ReLU activation function

```python
def leaky_relu(x):
    print(f"Leaky ReLU activation function: {max(0.01*x,x)}")
```

## Print the results

```python
sigmoid(sum_y)
step(sum_y)
tanh(sum_y)
sign(sum_y)
linear(sum_y)
relu(sum_y)
leaky_relu(sum_y)
```
